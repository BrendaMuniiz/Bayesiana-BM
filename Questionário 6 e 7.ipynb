## Funções útei para o projeto

```python
def amplitude(valor):
    minimo = min(valor)
    maximo = max(valor)
    return maximo - minimo
```

## Criando o dataset

Para a criação desse dataset, consideramos que estamos na cidade de São Carlos e desejamos estudar o tempo de entrega dos pedidos (em minutos) de um certo restaurante em um raio de até 2km. Além disso, todos os dados utilizados são fictícios.

Supomos que os dados possuem uma  distribuição normal com uma certa média, um certo desvio-padrão e com 30 amostras de tamanho 5:

```python
# Fixamos uma semente para reproduzir o estudo e obter os mesmos resultados
np.random.seed(0)
mu, sigma = 30, 5 # Média e desvio padrão (em minutos)
s = np.random.normal(mu, sigma, 150)

s = np. reshape(s, (30, 5))
s =pd.DataFrame(s)
s
```

```python
s.max(), s.min()
```

Em seguida, geramos mais 10 amostras de tamanho 5, pois supomos que 1 motoboy faltou no serviço. Dessa forma, o tempo de entrega aumentou de 30 para 40 minutos:

```python
mu, sigma = 40, 5 # # Média e desvio padrão (em minutos)
s2 = np.random.normal(mu, sigma, 50)
s2 = np. reshape(s2, (10, 5))
s2 =pd.DataFrame(s2)
s2
```

```python
s2.max(), s2.min()
```

```python
# Une os dois datasets
restaurante= s.append(s2)

# Arrumando o dataset do restaurante
restaurante.index = range(len(restaurante))
restaurante = restaurante.rename(columns={0: 'x1', 1: 'x2', 2: 'x3', 3: 'x4', 4: 'x5'})

restaurante
```

# Criando as colunas $\bar{X}$ e R

```python
restaurante['n_amostra'] = range(1, 41)
```

```python
restaurante['xbarra'] = restaurante.mean(axis=1)
```

```python
restaurante['R'] = restaurante.apply(amplitude, axis = 1)
```

```python
restaurante
```

```python
# Gerando o arquivo .csv
restaurante.to_csv('dados_restaurantes.csv',index=False, decimal='.', sep= ";")
```
